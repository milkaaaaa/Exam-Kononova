# Linear Regression

## Splitting

```python
a = ['R_D_Spend', 'Administration', 'Marketin_Spend', 'State']
```

```python
#Splitting the dataset into the Training set and Test set
X =df[a]
y = df['Profit']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Fitting Simple Linear Regression to the Training set
f1 = ['R_D_Spend']

from sklearn.linear_model import LinearRegression
sr = LinearRegression().fit(X_train[f1], y_train)

# Getting parameters
sr.coef_, sr.intercept_
```

```python
# Predicting the Test set results
y_pred = sr.predict(X_test[f1])

# Coefficient of determination R^2
sr.score(X_train[f1], y_train).round(3), sr.score(X_test[f1], y_test).round(3)
```

Получили высокие коэф. детерминации

```python
# Visualising the Test set results
plt.scatter(X_test['R_D_Spend'], y_test, color = 'red')
plt.plot(X_test['R_D_Spend'], sr.predict(X_test[f1]), 'bo')
plt.title('Profit vs R_D_Spend (Test set)')
plt.xlabel('R_D_Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# Visualising the Training set results
plt.scatter(X_train['R_D_Spend'], y_train, color = 'red')
plt.plot(X_train['R_D_Spend'], sr.predict(X_train[f1]), 'bo')
plt.title('Profit vs R_D_Spend (Training set)')
plt.xlabel('R_D_Spend')
plt.ylabel('Profit')
plt.show()
```

Видим, что полученная модель довольно хорошо описывает наши данные.
